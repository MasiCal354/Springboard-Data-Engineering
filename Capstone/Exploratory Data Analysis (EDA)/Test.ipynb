```python
df_trans = df.transpose()
new_header = df_trans.iloc[0] #grab the first row for the header
df_trans = df_trans[1:] #take the data less the header row
df_trans.columns = new_header #set the header row as the df header
df_trans = df_trans.groupby(level=0, axis=1).apply(lambda x: x.apply(list, axis=1))
df_trans = df_trans.reset_index(drop=True)
df_trans.columns.name = "ID"
df_trans = df_trans.reset_index()
df_trans = df_trans.set_index('index').apply(lambda x: x.apply(pd.Series).stack()).reset_index()
# df_trans.set_index(['index']).apply(pd.Series.explode).reset_index()
df_trans = df_trans.drop(columns=['CATEGORY', 'DEFINITION_XML', 'level_1', 'index'])
df_trans.to_csv('out.csv')
df_trans
```
