```python
import numpy as np
import pandas as pd
import json
from pandas_profiling import ProfileReport

with open('data.json') as data_json:
    data = json.load(data_json)

df = pd.json_normalize(data)
df
```

```python
df = pd.json_normalize(data['dataset'])
df
```

```python
df = pd.json_normalize(data['attribute'])
df
```

```python
df = pd.json_normalize(data['dimension'])
df
```

```python
df = pd.json_normalize(data['dimension'], record_path=['code'])
df
```

```python
df = pd.json_normalize(data['dimension'], record_path=['code',['attr']],)
pd.set_option('display.max_rows', df.shape[0]+1)
df
```

```python
df_trans = df.transpose()
new_header = df_trans.iloc[0] #grab the first row for the header
df_trans = df_trans[1:] #take the data less the header row
df_trans.columns = new_header #set the header row as the df header
df_trans = df_trans.groupby(lambda x:x, axis=1).aggregate(lambda x: ','.join(map(str, x)))
df_trans
```

```python
df_trans.sort_values('ds').groupby('ds')['fips'].apply(lambda df: df.reset_index(drop=True)).unstack()
df_trans.head()
```

```python
profile = ProfileReport(df, title="Pandas Profiling Report", explorative=True)
profile
```
